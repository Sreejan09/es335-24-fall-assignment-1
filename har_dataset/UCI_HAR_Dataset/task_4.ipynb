```python
import pandas as pd 
import numpy as np
import os
from sklearn.tree import DecisionTreeClassifier 
import matplotlib.pyplot as plt 
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix 
```

```python
# ------------ QUESTION 1 --------------
# ------- model creation -------- 
from MakeDataset import X_train,y_train, X_test
from sklearn.preprocessing import MinMaxScaler

X_train_total_acc = np.sum(X_train**2, axis=2)
X_test_total_acc = np.sum(X_test**2, axis=2)

model = DecisionTreeClassifier()
model.fit(X_train_total_acc, y_train)

scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(X_train_total_acc)
model_with_normalized_feature = DecisionTreeClassifier()
model_with_normalized_feature.fit(x_train_scaled, y_train)
```

```python
import pandas as pd
import os

ACTIVITIES = {
    1: 'WALKING',
    2: 'WALKING_UPSTAIRS',
    3: 'WALKING_DOWNSTAIRS',
    4: 'SITTING',
    5: 'STANDING',
    6: 'LAYING',
}

x_test_personal = []
y_test_personal = []

# Path to the main directory containing activity subfolders


for label, activity in ACTIVITIES.items():
    activity_folder = os.path.join('personal_data', activity)
    
    if os.path.exists(activity_folder):
        for filename in os.listdir(activity_folder):
            if filename.endswith('.csv'):
                file_path = os.path.join(activity_folder, filename)
                
                # Read 500 values from the 'aT' column, starting from offset 250
                data = pd.read_csv(file_path, skiprows=range(1, 250), nrows=500)
                data = data.iloc[:, -1:]
                data = data.T
                x_test_personal.append(data)
                y_test_personal.append(label)

# Concatenate all dataframes in x_test
x_test_personal = pd.concat(x_test_personal, ignore_index=True)

# Convert y_test to a pandas Series
y_test_personal = pd.Series(y_test_personal)

# Display the first few rows of x_test and y_test
print(x_test_personal.shape)
print(y_test_personal.shape)
```

```python
y_pred_personal = model.predict(x_test_personal)

# Calculate Accuracy
accuracy = accuracy_score(y_test_personal, y_pred_personal)
print(f"Accuracy: {accuracy:.2f}")

# Calculate Precision 
precision = precision_score(y_test_personal, y_pred_personal, average='weighted')
print(f"Precision : {precision:.2f}")

# Calculate Recall
recall = recall_score(y_test_personal, y_pred_personal, average='weighted')
print(f"Recall : {recall:.2f}")

# Confusion Matrix 
cm = confusion_matrix(y_test_personal, y_pred_personal)

print("Confusion Matrix")
print(cm)

```
